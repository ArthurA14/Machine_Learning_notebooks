```python
import numpy as np
import matplotlib.pyplot as plt
```

Régression avec SKLearn

```python
np.random.seed(0)
m = 100 # creating 100 samples
X = np.linspace(0 , 10, m).reshape(m, 1)
y = X + np.random.randn(m, 1)
```

```python
plt.scatter(X, y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(X, y) # Entraînement
model.score(X, y) # Evaluation - Ici, on calcule le coefficient de détermination R2
predictions = model.predict(X) # Utilisation / Nouvelles prédictions
print(predictions[:10])
```

```python
plt.scatter(X, y)
plt.plot(X, predictions, c='red')
```
